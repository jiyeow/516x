####  Naive Bayes Classification

Use [Naive Bayes Classification](https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.GaussianNB.html) to create a model to classify wine base on attributes.  Justify how good the model is for the wine classification.  Note that some of the metrics we've used in class are only for *binary* classifications, so may not eb applicable here.


```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC, LinearSVC
from sklearn.metrics import classification_report, f1_score, accuracy_score, confusion_matrix
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold, cross_val_score, train_test_split 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import learning_curve
```

```python
# split dataset into training and testing
X = df.loc[:, df.columns != 'Class']
y = df['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(X.shape)
```

```python
nb = MultinomialNB()
nb.fit(X_train, y_train)
```

```python
y_pred_class = nb.predict(X_test)
print(y_pred_class)
```

```python
# calculate accuracy of class predictions
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_class)
```

```python
# same steps as above, but only using selected parameters (instead of all parameters) to predict wine class
X = df[['Alcohol', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Color intensity']]
y = df['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
nb = MultinomialNB()
nb.fit(X_train, y_train)
y_pred_class = nb.predict(X_test)
metrics.accuracy_score(y_test, y_pred_class)
```

### Model developed using selected parameters 'Alcohol', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Color intensity' had higher accuracy than using all parameters

```python
# print the confusion matrix
cm = metrics.confusion_matrix(y_test, y_pred_class)
cmdf = pd.DataFrame(cm,index=['1','2','3'], columns=['1','2','3'])
print("The confusion matrix looks like following...\n")
cmdf
```

```python
print("The classification report is as follows...\n")
print(classification_report(y_pred_class,y_test))
```
